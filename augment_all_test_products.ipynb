```python
import production.process_llm as llm
import time
import pandas as pd

```

```python
df = pd.read_csv('data/all_scanned_receipts_cleaned.csv', index_col=0)
```

```python
df
```

```python
augment_data = []
```

```python
categories = llm.get_rewe_categories()
```

```python
for product_abbr in df.product_abbr:
    print(product_abbr)

    try:
        response = llm.process_abbr_item(product_abbr, categories)
        augment_data.append(response)
    except:
        print('Error', product_abbr)
    time.sleep(12)
    print('**sleeping')
```

```python
augment_data
```

```python
processed_data = pd.DataFrame(augment_data).drop('product_abbr', axis=1).join(df, how='inner')
processed_data
```

```python
pd.DataFrame(augment_data).to_csv('data/all_scanned_receipts_processed_missings.csv')
```
