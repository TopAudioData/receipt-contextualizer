```python
with open('../data/allcategories.html', 'r') as f:
    soup = BeautifulSoup(f, "html.parser")

```

```python
soup
```

```python
lists = soup.find_all("li", class_="lr-category__item")
```

```python
lists_top = soup.find_all("li", class_="lr-category__top lr-category__item")
```

```python
lists_top
```

```python
categories_all = []
for list in lists:
    category = list.find('a').text
    print(category)
    categories_all.append(category)
```

```python
categories_sub = [cat for cat in categories_all if not cat.startswith('Alle ')]
```

```python
categories_top = []
for list in lists_top:
    cat = list.find('a').text
    print(cat)
    categories_top.append(cat)
```

```python
categories_top
```

```python
categories_sub
```

```python
categories_sub = [cat for cat in categories_sub if not cat in categories_top]
```

```python
cat_nums = [cat for cat in categories_all if cat.startswith('Alle ')]
```

```python
import re

cat_nums[0]

re.findall(cat_nums[0], '"\d"g')
```

```python
numbers = []

for cat_num in cat_nums:
    match = re.search(r'\d+', cat_num)
    if match:
        numbers.append(int(match.group()))

print(numbers)
```

```python
categories_dict = dict()

counter = 0

for i, num in enumerate(numbers):
    cat_top = categories_top[i]
    print(cat_top)
    
    cat_sub = categories_sub[counter:counter+num]
    print(cat_sub)

    categories_dict[cat_top] = cat_sub

    counter += num

"""
    cat_sub = categories_sub[counter:cat_num[counter]]
    for cat_sub in categories_sub:
        print('',cat_sub)"""
```
